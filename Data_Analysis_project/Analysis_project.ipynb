```python
import pandas as pd
```

```python
df = pd.read_csv("IEEE ZSB Database'24.xlsx - Form Responses 1.csv")
```

```python
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df
```

```python
print(df.columns)
```

```python
def parse_date(value):
    try:
        return pd.to_datetime(value, format='%m/%d/%Y', errors='raise')
    except (ValueError, TypeError):
        if str(value).isdigit():
            return pd.to_datetime(value, format='%Y', errors='coerce')
        return pd.NaT

df['Date of birth '] = df['Date of birth '].apply(parse_date)

print(df['Date of birth '].head(40))
```

```python
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
from datetime import datetime

current_year = datetime.now().year

df['Year of birth'] = df['Date of birth '].dt.year
df['Age'] = current_year - df['Year of birth']

df[['Date of birth ', 'Age']].head()
```

```python
df.rename(columns={"If you're part of two teams, what's of your second team?": 'Second Team'}, inplace=True)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Committee\\Chapter', data=df)
plt.xticks(rotation=45, ha='right')
```

```python
df = df[(df['Age'] >= 18) & (df['Age'] < 100)]
```

```python
problematic_rows = df[(df['Age'] < 0) | (df['Age'] < 10)]
print(problematic_rows)
```

```python
sns.countplot(x='Age', data=df)
```

```python
def categorize_faculty(faculty_name):
    if 'engineering' in faculty_name.lower():
        return 'Engineering'
    elif 'computer' in faculty_name.lower() or 'informatics' in faculty_name.lower() or 'Computer'in faculty_name.lower()or 'fci'in faculty_name.lower():
        return 'Computer Science'
    elif 'art' in faculty_name.lower() :
        return 'Arts'
    elif 'commerce' in faculty_name.lower() :
        return 'Commerce'
    else:
        return 'Other'

df.loc[:, 'Faculty Category'] = df['Faculty'].apply(categorize_faculty)

plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Faculty Category', data=df)

for bar in ax.patches:
    bar.set_width(0.5) 

plt.xticks(rotation=45, ha='right')
plt.xlabel('Faculty Category')
plt.ylabel('Count')
plt.title('Count of Faculty Categories')

plt.show()
```
